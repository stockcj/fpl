```python
from selenium import webdriver
import pandas as pd
import lxml.html
import re
```

```python
options = webdriver.ChromeOptions()
driver = webdriver.Chrome(chrome_options=options)
driver.get('https://understat.com/league/EPL')
source = driver.page_source
tree = lxml.html.fromstring(source)
```

```python
header_els = tree.xpath('//div[@id="league-chemp"]//table//th')
headers_list = []
for header in header_els:
    headers_list.append(header.text_content())
print(headers_list)
```

```python
tbody = tree.xpath('//div[@id="league-chemp"]/table/tbody')[0]
rows = []
for tr in tbody.findall('tr'):
    row = tr.findall('td')
    current_row = []
    for item in row:
        current_row.append(item.text_content())
    rows.append(current_row)
```

```python
xg_df = pd.DataFrame(rows, columns=headers_list)
xg_df['xG'] = xg_df['xG'].apply(lambda x: re.split('[+-]',x)[0])
xg_df['xGA'] = xg_df['xGA'].apply(lambda x: re.split('[+-]',x)[0])
xg_df['xPTS'] = xg_df['xPTS'].apply(lambda x: re.split('[+-]',x)[0])
xg_df
```

```python
xg_df['xG'] = pd.to_numeric(xg_df['xG'])
xg_df['xGA'] = pd.to_numeric(xg_df['xGA'])
xg_df['xPTS'] = pd.to_numeric(xg_df['xPTS'])
xg_df['G'] = pd.to_numeric(xg_df['G'])
xg_df['GA'] = pd.to_numeric(xg_df['GA'])
xg_df['GD'] = xg_df['G']-xg_df['GA']
xg_df
```

```python
cols = list(xg_df.columns.values)
print(cols)
```

```python
cols = ['№', 'Team', 'M', 'W', 'D', 'L', 'G', 'GA', 'GD', 'PTS', 'xG', 'xGA', 'xPTS']
xg_df = xg_df.reindex(columns=cols)
xg_df
```
